```python
## Path to csv files
path = r'D:\Koodit\Ekonometria\Econometrics\data' # use your path
all_files = glob.glob(os.path.join(path , "*.csv"))
li = []

# Open csv files using pandas 
for filename in all_files[2:]:
    df = pd.read_csv(filename, index_col=None, header=0,sep=';',encoding = "utf-8")
    li.append(df)
    
# Concat dataframes into one
frame1 = pd.concat(li, axis=0, ignore_index=True)
li=[]
for filename in all_files[:2]:
    df = pd.read_csv(filename, index_col=None, header=0,sep=';',encoding = "ISO-8859-1")
    li.append(df)

frame2 = pd.concat(li, axis=0, ignore_index=True)
```

```python
all_data3=pd.concat([helsinki,tampere])
```

```python
all_data3['Helsinki'] = np.where(all_data3['koulun_nimi'].isin(helsingin_lukiot), 1, 0)
```

```python
all_data3['Before'] = np.where(all_data3['tutkintokerta'].isin(['2017K','2017S','2018K','2018S','2019K','2019S']), 1, 0)
all_data3['After'] = np.where(all_data3['tutkintokerta'].isin(['2021K','2021S','2022K','2022S']), 1, 0)
```

```python
all_data3['interaction'] = all_data3.Helsinki * all_data3.After
```

```python
all_data3
```

```python
from statsmodels.formula.api import ols
ols = ols('yht ~ Helsinki + After + interaction', data=all_data3).fit()
print(ols.summary())
```
