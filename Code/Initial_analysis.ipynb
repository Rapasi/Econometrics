```python
import pandas as pd
import matplotlib.pyplot as plt
import glob
import os
```

```python
## Path to csv files
path = r'D:\Koodit\Ekonometria\Econometrics\data' # use your path
all_files = glob.glob(os.path.join(path , "*.csv"))
li = []

# Open csv files using pandas 
for filename in all_files:
    df = pd.read_csv(filename, index_col=None, header=0,sep=';',encoding='UTF8')
    li.append(df)
# Concat dataframes into one
frame = pd.concat(li, axis=0, ignore_index=True)
```
